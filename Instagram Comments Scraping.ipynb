```python
import time
import pandas as pd
from selenium import webdriver
```

### Code

```python
def scrape_instagram_comments(url):
    """
    Extracts the comments from the instagram post given by the URL.
    Args:
        url (str): The URL to the instagram post
    Return:
        dataframe of users and comments
    """
    
        # Asking for username and password for a real instagram account
    insta_user = input("Enter your instagram account username: ")
    insta_pass = input("Enter the login password: ")
    
        # Note: Download and replace the chromedriver path if it's different
    driver = webdriver.Chrome('C:/Program Files/Google/Chrome/Application/chromedriver.exe')

        # Open Instagram website, maximizes the current window, and wait 2 seconds for page to load.
    driver.get('https://www.instagram.com/')
    driver.maximize_window()
    time.sleep(2)

        # Login steps
    username = driver.find_element_by_css_selector("input[name='username']") 
    password = driver.find_element_by_css_selector("input[name='password']") 
    username.clear()
    password.clear()
    username.send_keys(insta_user)
    password.send_keys(insta_pass)
    driver.find_element_by_css_selector("button[type='submit']").click()
    time.sleep(5)

        # Navigates to the URL, maximizes the current window, and wait 5 seconds for page to load.
    driver.get(url)
    time.sleep(5)

        # Define a function that check if there's "Load More Replies" burron
    def check_exists_by_xpath(xpath):
            try:
                driver.find_element_by_xpath(xpath)
                return True
            except:
                return False

        # Load more replies as exists
    while check_exists_by_xpath("//div/ul/li/div/button"):
            load_more_comments_element = driver.find_element_by_xpath("//div/ul/li/div/button")
            load_more_comments_element.click()
            time.sleep(5)

        # Extracting all users and comments and sort them in lists.
    user_names = []
    user_comments = []
    comments = driver.find_elements_by_class_name('gElp9')
    for comment in comments:
            container = comment.find_element_by_class_name('C4VMK').text.split('\n')
            name = container[0]
            if container[1]=='Verified':
                content = container[2]
            else:
                content = container[1]

            user_names.append(name)
            user_comments.append(content)

        # Save data in Dataframe and show it.    
    df = pd.DataFrame({'user_name' : user_names, 'user_comment' : user_comments})

         # Close the browser
    driver.close()
    
    return df
```

### Test the code
